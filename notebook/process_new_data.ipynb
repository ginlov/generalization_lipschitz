```python
variance = data.groupby(by=["model", "num_cluster"])[["g_value 0.01 mean", "g_value 0.05 mean", "g_value 0.1 mean", "rob", "localrob", "localavg"]].var()
```

```python
mean = data.groupby(by=["model", "num_cluster"])[["g_value 0.01 mean", "g_value 0.05 mean", "g_value 0.1 mean", "rob", "localrob", "localavg"]].mean()
```

```python
for i in range(20, 24): 
    var = list(variance.iloc[i].values)
    me = list(mean.iloc[i].values)
    list_of_items = []
    for i in range(len(var)):
        list_of_items.append(me[i])
        list_of_items.append(var[i])

    print("{:1.3f}$\pm${:2.2E} & {:1.3f}$\pm${:2.2E} & {:1.3f}$\pm${:2.2E} & {:1.3f}$\pm${:2.2E} & {:1.3f}$\pm${:2.2E} & {:1.3f}$\pm${:2.2E} \\\\".format(*list_of_items))
```
